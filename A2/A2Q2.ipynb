```python
import pandas as pd
from matplotlib import pyplot as plt
# Independently designed by Shao Shi
# Data filtering method: remove -9999 values of temperature,
# and the air temperature quality code should be 1 at the same time

raw_data = pd.read_csv("Baoan_Weather_1998_2022.csv", header=0)
TMP = raw_data["TMP"].str.split(',', expand=True).astype(int)
time = raw_data["DATE"].str.split('-', expand=True)
TMP["Month"] = time[1]
TMP["Year"] = time[0]
TMP = TMP.loc[(TMP[0] != +9999) & (TMP[1] == 1)]
ave = TMP.groupby(['Year', 'Month']).mean()[0]/10
ave.plot()
plt.show()
```
